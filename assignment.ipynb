1. A model is linear if it is linear in its parameters. This means that the predicted outcome is a linear combination of the parameters (i.e. the coefficients) multiplied by the predictor variables. In other words, the coefficients are additive and not involved in exponentiation, multiplication with each other, or complex transformations. 
2. The interpretation of a dummy variable's coefficient depends on the model's intercept. When there is an intercept, one category is excluded to act as a baseline, and each dummy coefficient is the difference in the dependent variable compared to this baseline. When there is no intercept, all categories are included, and each coefficient directly represents the expected outcome for that category. Thus, the intercept determines whether dummy coefficients indicate a difference from the baseline or an absolute value.
3. Linear regression is not ideal for classification because it predicts continuous outcomes rather than categorical outcomes. Linear regression produces continuous values that can fall outside the range of valid class probabilities (e.g. -1 or 2), which don't make sense for classification. It also assumes a linear relationship between predictors and the target, which doesn't align with most classification tasks, where we want discrete class assignments rather than continuous predictions.
4.  A sign that a model is overfitting is if there are a large number of features, which could cause the model to focus too much on specific patterns in the training set, leading to overfitting. Another sign is if the model has a high accuracy on the training data but poor accuracy on the test data, which shows that the model has memorized the training set rather than the overall patterns. Another sign is if the model's predictions are highly sensitive to changes in the test data, which shows that the model has adapted too closely to the specific data in the training set. 
5. Collinearity is when two or more predictor variables are highly correlated, which makes it difficult to distinguish their individual effects on the outcome variable. The two-stage least squares technique address this. In the first step, the problematic variable is regressed on instrumental variables, which are the predictors that are related to the collinear variable but not to the error term. In the second step, this predicted value replaces the original variable in the main regression, which will provide more reliable coefficient values. By doing this, each predictor's effect is more isolated, which leads to a more accurate model.
6. To incorporate nonlinear relationships between y and x, one could use polynomial features, which involves adding terms like x^2, x^3, and so forth to capture curves in the data without changing the model's structure. One could also use transformations, such as applying logs, square roots, or exponential functions to x or y to linearize certain nonlinear relationships. For example, if y grows exponentially with x, one could take the log of y or x to make the relationship linear. 
7. In linear regression, the intercept is the expected value of the outcome variable y when all predictors are zero (i.e. when the predictors exhibit no influence). A slope coefficient for a variable indicates the expected change in y for a one-unit increase in that predictor, when all other variables are constant. In other words, it represents how sensitive y is to changes in that predictor. For a dummy variable, the coefficient represents the difference in y between the group represented by the dummy variable and the baseline/reference group, assuming there's an intercept in the model. If the coefficient is positive, it indicates that the dummy group has a higher y value than the reference group.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
# Part 1 - Compute the average prices and scores by Neighbourhood; which borough is the most expensive on average? Create a kernel density plot of price and log price, grouped by Neighbourhood

df = pd.read_csv('./data/Q1_clean.csv')
avg_price = df.groupby('Neighbourhood ')['Price'].mean()
avg_score = df.groupby('Neighbourhood ')['Review Scores Rating'].mean()
print(avg_price, '\n')  # Manhattan is the most expensive borough on average, with an average price of $183 per night
print(avg_score)

sns.kdeplot(data=df, x='Price', hue='Neighbourhood ')
plt.show()
sns.kdeplot(data=df, x=np.log(df['Price']), hue='Neighbourhood ')
plt.show()
```

```python
# Part 2 - Regress price on `Neighbourhood ` by creating the appropriate dummy/one-hot-encoded variables, without an intercept in the linear model and using all the data. Compare the coefficients in the regression to the table from part 1. What pattern do you see? What are the coefficients in a regression of a continuous variable on one categorical variable?

y = df['Price']  # Dependent variable
X = pd.get_dummies(df['Neighbourhood '], dtype='int')  # Independent variable
reg = LinearRegression(fit_intercept=False).fit(X, y)

results = pd.DataFrame({'Neighbourhood': reg.feature_names_in_, 'Coefficient': reg.coef_})
print(results)  # The coefficients by neighborhood in the regression are the same as the average prices by neighborhood in Part 1. In linear regression, each coefficient in a model without an intercept where the predictor variable is categorical represents the mean outcome for each level of the categorical variable. This is because there is no baseline from which adjustments are made.
```

```python
# Part 3 - Repeat part 2, but leave an intercept in the linear model. How do you have to handle the creation of the dummies differently? What is the intercept? Interpret the coefficients. How can I get the coefficients in part 2 from these new coefficients?

X = pd.get_dummies(df['Neighbourhood '], dtype='int', drop_first=True)
reg = LinearRegression().fit(X, y)

results = pd.DataFrame({'Neighbourhood': reg.feature_names_in_, 'Coefficient': reg.coef_})
print(results, '\n')
print(reg.intercept_)  # I adjusted the creation of the dummy variables by omitting the Bronx from the neighborhood categories. The intercept becomes the average price for homes in the Bronx, 75.27649769584711, and each coefficient represents the difference in average price between homes in each neighborhood and homes in the Bronx. To yield the coefficients from the previous regression, add each neighborhood's coefficient from this regression to the intercept.
```

```python
# Part 4 - Split the sample 80/20 into a training and a test set. Run a regression of `Price` on `Review Scores Rating` and `Neighbourhood `. What is the R^2 and RMSE on the test set? What is the coefficient on `Review Scores Rating`? What is the most expensive kind of property you can rent?

from sklearn.model_selection import train_test_split

X = df.loc[:, ['Review Scores Rating', 'Neighbourhood ']]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=100)
Z_train = pd.concat([X_train['Review Scores Rating'], pd.get_dummies(X_train['Neighbourhood '], dtype='int')], axis=1)
Z_test = pd.concat([X_test['Review Scores Rating'], pd.get_dummies(X_test['Neighbourhood '], dtype='int')], axis=1)

reg = LinearRegression(fit_intercept=False).fit(Z_train,y_train)

y_hat = reg.predict(Z_test)
print('R2: ', reg.score(Z_test, y_test))
rmse = np.sqrt(np.mean((y_test - y_hat)**2))
print('RMSE: ', rmse, '\n')

results = pd.DataFrame({'Variable':reg.feature_names_in_, 'Coefficient': reg.coef_})
print(results, '\n')
print(100 * 1.032257 + 89.421020)  # The most expensive property you can rent is a 100-rated place in Manhattan for about 192.65 per night
```

```python
# Part 5 - Split the sample 80/20 into a training and a test set. Run a regression of `Price` on `Review Scores Rating` and `Neighbourhood ` and `Property Type`. What is the R^2 and RMSE on the test set? What is the coefficient on `Review Scores Rating`? What is the most expensive kind of property you can rent?

y = df['Price']
X = df.loc[:,['Review Scores Rating', 'Neighbourhood ', 'Room Type'] ]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=100)
Z_train = pd.concat([X_train['Review Scores Rating'], 
                    pd.get_dummies(X_train['Neighbourhood '], dtype='int'),
                    pd.get_dummies(X_train['Room Type'], dtype='int')],
                    axis=1)
Z_test = pd.concat([X_test['Review Scores Rating'], 
                    pd.get_dummies(X_test['Neighbourhood '], dtype='int'),
                    pd.get_dummies(X_test['Room Type'], dtype='int')],
                    axis=1)

reg = LinearRegression(fit_intercept=False).fit(Z_train,y_train)

y_hat = reg.predict(Z_test)
print('R2: ', reg.score(Z_test, y_test))
rmse = np.sqrt(np.mean((y_test - y_hat)**2))
print('RMSE: ', rmse, '\n')
results = pd.DataFrame({'Variable':reg.feature_names_in_, 'Coefficient': reg.coef_})
print(results, '\n')
print(100 * 0.626912 + 110.617820 + 53.693304)  # The most expensive property you can rent is a 100-rated entire home/apartment in Manhattan for about 227.00 per night
```

**Q4.** This is a math question to review the derivation of the OLS estimator (but only if you are into that kind of thing!). We are going to do it slightly differently from what we did in class, though. We will use a linear predictor and minimize the Sum of Squared Errors, just as in class. But, we are going to de-mean $X$ first, creating another variable $z_i = x_i - \bar{x}$ where
$$
\bar{x} = \dfrac{1}{N} \sum_{i=1}^N x_i,
$$
so the model is $\hat{y}_i = a + b z_i$ and the `SSE` is
$$
\text{SSE}(a,b) = \sum_{i=1}^N (y_i - a - bz_i)^2.
$$

  1. Take partial derivatives of the `SSE` with respect to $a$ and $b$. You should get

\begin{alignat*}{3}
\sum_{i=1}^N -2(y_i - a- bz_i) &=& 0 \\
\sum_{i=1}^N -2(y_i - a - bz_i)z_i &=& 0.
\end{alignat*}

  2. Solve for the solutions to the above equations. Big hint: $\bar{z} = 0$, since we subtracted the mean of $x$ from $x$ to get $z$. You should get

\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\sum_{i=1}^N(y_i - \bar{y})z_i}{\sum_{i=1}^N z_i^2}.
\end{alignat*}

  3. Substitute $z_i = x_i - \bar{x}$ back into the above equations. You should get
  
\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\sum_{i=1}^N(y_i - \bar{y})(x_i-\bar{x})}{\sum_{i=1}^N (x_i-\bar{x})^2},
\end{alignat*}

which can be written in terms of sample covariance and sample variance as:

\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\text{cov}(x,y)}{\text{var}(x)}.
\end{alignat*}

This is typically the preferred way of expressing the OLS coefficients.

5. Suppose you have measurement error in $x$ which artificially inflates its variance (e.g. bad data cleaning). What happens to the $b^*$ coefficient? How will affect your ability to predict? (This phenomenon is called **attenuation**.)
